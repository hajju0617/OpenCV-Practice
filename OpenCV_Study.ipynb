```python
import cv2
import matplotlib.pyplot as plt

img_path = '/content/drive/MyDrive/Cat.png'
img = cv2.imread(img_path)

if img is None:
    print(f"이미지를 불러오지 못했습니다: {img_path}")
else:
    # 전처리 및 컬러 변환 (OpenCV BGR -> Matplotlib RGB)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 어둡게 (-50)
    dark = cv2.subtract(img, (50, 50, 50, 0))
    dark_rgb = cv2.cvtColor(dark, cv2.COLOR_BGR2RGB)

    # 블러링
    blurred_rgb = cv2.GaussianBlur(img_rgb, (15, 15), 0)

    # 시계 방향으로 90도 회전
    rotated_rgb = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)

    # 엣지 추출
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 200)


    titles_2x2 = ['Original', 'Darker', 'Blurred', 'Rotated']
    images_2x2 = [img_rgb, dark_rgb, blurred_rgb, rotated_rgb]

    plt.figure(figsize=(10, 8))

    for i in range(4):
        # subplot(행의 개수, 열의 개수, 순서)
        plt.subplot(2, 2, i + 1)
        plt.imshow(images_2x2[i])
        plt.title(titles_2x2[i], fontsize=12)
        # x, y축 눈금 없앰.
        plt.axis('off')

    # 이미지 간의 간격과 제목이 겹치지 않도록 여백을 자동으로 최적화.
    plt.tight_layout()
    plt.show()

    # Edges 단독 배치
    print("\n" + "="*50 + "\n")

    plt.figure(figsize=(10, 8))
    # cmap='gray': 명확한 흑백 출력을 위해 지정.
    plt.imshow(edges, cmap='gray')
    plt.title('Edges (Canny)', fontsize=12)
    plt.axis('off')

    plt.show()
```
